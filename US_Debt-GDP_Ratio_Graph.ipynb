```python
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.width',250)
pd.set_option('display.max_columns',15)
pd.set_option('display.min_rows', 50)

df0 = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/04 Week 4: Explanatory Viz/US GDP vs Debt.xlsx')
print(df0.shape)
print(df0.head(2))
```

```python
#Data Preparation
df1 = df0.dropna(how = 'any')
quarter = df1['Quarter'].values
gdp = df1['Gross Domestic Product ($mil)'].values
debt = df1['Total Public Debt ($mil)'].values
debtratio = debt/gdp
```

```python
#Graph

plt.style.use('seaborn')
plt.plot(quarter, debtratio)
plt.xticks(rotation = 45)
plt.xlabel("Years in which U.S. carried debt", fontsize = 18);
plt.ylabel('Debt to GDP Ratio', fontsize = 18);
plt.title('Debt Growing Nonproportional to GDP', fontsize=20)
```
